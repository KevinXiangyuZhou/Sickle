```python
import numpy as np
import matplotlib.pyplot as plt
import copy

def barchart(data_1, data_2, x_label, y_label, N):
    # plt.yscale('log')
    ind = np.arange(N) 
    width = 0.35       
    plt.bar(ind, data_1, width, label="default")
    plt.bar(ind + width, data_2, width,
        label='with_analysis')

    plt.ylabel(y_label)
    plt.title(f"{x_label} vs {y_label}")

    plt.xlabel(x_label)
    plt.xticks(ind + width / 2, [str(i) for i in range(1, N + 1)])
    plt.legend(loc='best')
    plt.show()
    
def linechart(data_1, data_2, x_label, y_label, N):
    plt.yscale('log')
    ind = np.arange(N) 
    plt.plot(ind, data_1, '-')
    plt.plot(ind, data_2, '-')
    plt.xlabel(x_label)
    plt.ylabel( y_label)
    plt.title(f"{x_label} vs {y_label}")
    plt.legend(['default', 'with_analysis'])
    plt.show()

```
